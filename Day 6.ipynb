# MEAN,MEDIAN and MODE

# MEAN

```python
import statistics
marks = [45, 50, 55, 60, 65]
mean_value = statistics.mean(marks)

print("Mean =", mean_value)
```

# MEDIAN

```python
import statistics
marks = [45, 50, 55, 60, 65]
median = statistics.median(marks)
print("Median:",median)
```

# MODE

```python
import statistics
data = [10, 20, 30, 40, 50]
mode = statistics.mode(data)
print("Mode:",mode)
```

# MEAN,MEDIAN and MODE using Pandas

# PROBABILITY

```python
sample_space = {1,2,3,4,5,6}
event_A = {2,4,6}
print('Sample Space(S):',sample_space)
print("Event A:",event_A)

#2. Probability P(A)
P_A = len(event_A)/ len(sample_space)
print('\nProbability of Event A')
print('P(A) =', P_A)

#3. Conditional Probability
# P(A/B)
event_B = {4,5,6} # Numbers greater than 3
intersection = event_A.intersection(event_B)
P_A_given_B = len(intersection)/ len(event_B)
print('\nEvent B:', event_B)
print('A ∩ B:',intersection)
print('P(A|B) =', P_A_given_B)

#4. Baye's Theorem
P_Disease = 0.01
P_Positive_given_Disease = 0.99
P_Positive = 0.05
P_Disease_given_Positive = (P_Positive_given_Disease * P_Disease )/ P_Positive
print("\nBaye's Theorem Example")
print("P(Disease) =", P_Disease)
print("P(Positive|Disease) =", P_Positive_given_Disease)
print("P(Positive) =", P_Positive)
print("P(Disease|Positive) =", round(P_Disease_given_Positive,4))
print("Percentage =", round(P_Disease_given_Positive * 100, 2),"%")
```

# CORRELATION

```python
#Pearson Correlation
import pandas as pd
df = pd.DataFrame({
    'Hours': [1,2,3,4,5],
    'Marks': [20,40,50,70,90]
})
corr = df['Hours'].corr(df['Marks'])
print("Pearson Correlation:",corr)

#Spearman Correlation
import pandas as pd
df = pd.DataFrame({
    'Rank_Study': [1,2,3,4,5],
    'Rank_Marks': [2,1,3,5,4]
})
corr = df['Rank_Study'].corr(
    df['Rank_Marks'],
    method = 'spearman'
)
print("Spearman Correlation:", corr)
# Correlation
import pandas as pd
df = pd.DataFrame({
    'Hours': [1,2,3,4,5],
    'Marks': [20,40,50,70,90],
    'Sleep': [8,7,6,5,4]
})
print(df.corr())

#Heatmap Correlation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    'Hours': [1,2,3,4,5],
    'Marks': [20,40,50,70,90],
    'Sleep': [8,7,6,5,4]
})
sns.heatmap(df.corr(),
            annot=True,
            cmap='plasma')
plt.show()
#Common cmap values
#coolwarm,viridis,YlGnBu,blues,red,greens,magma
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Generate normaally distributed data
data = np.random.normal(
    loc = 70,
    scale = 10,
    size = 1000
)
#Calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data)
print("Mean:", round(mean, 2))
print("Standard Deviation:", round(std, 2))
#Plot histrgame
plt.figure(figsize=(8,5))
sns.histplot(data, bins=30,kde=True)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
