```python
import pandas as pd
a = pd.Series([1,2,3,4,5,6])
b = pd.Series([True,False,True])
c = pd.Series(['varshine','varshini','riya','nila'])
print(a)
print(b)
print(c)
print(a.values)
print(a.index)
print(a.dtype)
# Accessing elements
print(a[1])
# Accessing elements using loc
print(c.loc[1])
print(a.iloc[1])
# Slicing
print(c[1:4])
print(c.loc[1:3])
print(c.iloc[2:4])
```

```python
import pandas as pd
arr = pd.Series([1,2,3,4,5,6],index=['A','B','C','D','E','F'])
print(arr)
print(arr[['A','C','E']])
print(arr['E'])
# Vectorized Operation - no loop required
print(arr+5)
print(arr.iloc[1])
print(arr.loc['A'])
# Statistical function
print(arr.sum())
print(arr.std())
print(arr.max())
print(arr.min())
print(arr.mean())
print(arr.median())
# Boolean indexing
print(arr[arr>30])
# Count values
print(len(arr))
#it return individual value count
print(arr.value_counts())
#String operation
name = pd.Series(['priya','Nila','Ravi','rahul','raji'])
print(name.str.upper())
print(name.str.len())
```

```python
import pandas as pd
# Method1 : From dictionary
data = {
    'name':['varshine','ravi','rahul','raj'],
    "rollno":[27,20,24,48],
    'Age':[19,20,23,25]}
df = pd.DataFrame(data)
print(df)
# From Series to dataframe
s1 = pd.Series([1,2,3,4,5,6])
s2 = pd.Series([5.4,6.7,3.4,9.8,7.9,8.0])
s3 = pd.Series([True,False,True,False])
df = pd.DataFrame({'SNo':s1,
                   'CGPA':s2,
                   'Attendance':s3})
print(df)
# List of list
data = [['Nila' ,20,98],
    ['Nikila' ,19,99],
    ['Adhi' ,21,97]]
df = pd.DataFrame(data,columns = ["name","Age","Mark1"])
print(df)
# List of Dictionary
data = [
    {"Name":"varshine","Course":"AIDS","Grade":'A'},
    {"Name":"varsh","Course":"AIDS","Grade":'A+'},
    {"Name":"shine","Course":"AIDS","Grade":'A++'}]
df = pd.DataFrame(data,index = ['std1','std2','std3'])
print(df)
# Adding one extra column
df['Marks'] = [98,87,67]
print(df)
# Adding one extra row
std4 = pd.DataFrame({"Name":"nila","Course":"AIDS","Grade":'O',"Marks":99},
                   index=['std4'])
df = pd.concat([df,std4])
print(df)
# Create multiple rows
students = pd.DataFrame([{"Name":"Nila","Course":"AIDS","Grade":'O',"Marks":99},
                         {"Name":"Nikila","Course":"AIDS","Grade":'A++',"Marks":98}], index = ["std5","std6"])
df = pd.concat([df,students])
print(df)
# Column Access by bracket notation and dot notation
print(df.Course)
print(df[["Grade","Marks"]])
# Shape attribute
print(df.shape)
# Find no of rows
print(df.shape[0])
# Find no of columns
print(df.shape[1])
# Dtype attribute
# Used to check data type of all columns
print(df.dtypes)
# Info
print(df.info())
print(df.describe())
# Select speific row
print(df.loc[['std1','std4',]])
print(df.iloc[3])
# Update value
df.loc['std3','grade'] = 'B++'
print(df)
# Delete column
df.drop('grade', axis = 1,inplace = True)
print(df)
# delete row
df = df.drop('std6', axis = 0)
print(df)
# Sorting
df.sort_values('Marks', inplace = True)
print(df)
df.sort_values('Marks', ascending = False,inplace = True)
print(df)
# Stastical function for particular column
print(df['Marks'].max())
```
