```python
import pandas as pd
df = pd.read_json("titanic.json")
print(df)
df.head()
# Checks missing values
#isnull()
# Used to identify missing values
print(df.isnull())
# Count missing values for all field
df.isnull().sum()
# Find rows having missing age
df[df['Age'].isnull()]
print(df.columns)
# Fillna()
# Replace missing values
# Fill Age with mean
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
# Fill embarked with most frequent value
# by using mode() means the most frequently occuring value in a column
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_embarked)
#dropna()
# Remove missing data
# Remove rows
df.dropna()
# Remove column
df.dropna(axis = 1)
# Groupby
# Average Fare per class
print(df.groupby('Pclass')['Fare'].mean())
# Count survivors by gender
df.groupby('Sex')['Survived'].sum()
# Multiple Aggrigations
# agg()
print(df.groupby('Pclass').agg({
    'Age':['mean','min','max'],
    'Fare':['mean','sum']
}))
# Interpolate()
# Estimate missing values
sample = pd.DataFrame({
    'Age':[20,25,None,35,40]
})
sample['Age'] = sample['Age'].interpolate()
print(sample)
# Transform()
# Returns same number 
```

# EXCEL FILE OPERATIONS

```python
import pandas as pd
students = pd.read_excel("Merge_Join_Concat_Dataset_50_Records.xlsx", sheet_name = "students")
marks = pd.read_excel("Merge_Join_Concat_Dataset_50_Records.xlsx", sheet_name = "marks")
contact = pd.read_excel("Merge_Join_Concat_Dataset_50_Records.xlsx", sheet_name = "contact")
print(students)
print(marks)
print(contact)

#1.INNER MERGGE
#Return only matching StudentIDs from both dataframes
inner_df = pd.merge(
    students,
    marks,
    on = "Student_ID",
    how = "inner"
)
print(inner_df.head())
print(inner_df.shape)

#2. LEFT MERGE
# Keep all records from the left DataFrame
left_df = pd.merge(
    students,
    marks,
    on = "Student_ID",
    how = "left"
)
print(left_df.tail())
# Even if marks are missing, students remain

# 3. RIGHT MERGE
right_df = pd.merge(
    students,
    marks,
    on = "Student_ID",
    how = "right"
) 
print(left_df.tail())

# 4. OUTER MERGE
outer_df = pd.merge(
    students,
    marks,
    on = "Student_ID",
    how = "outer"
)
print(outer_df.head())
print(outer_df.shape)
#Combine Students + Marks + Contact
student_complete = pd.merge(
    students,
    marks,
    on = "Student_ID",
    how = "left"
)
student_complete = pd.merge(
    student_complete,
    marks,
    on = "Student_ID",
    how = "left"
)
print(student_complete.head())

# 6. CONCAT ROW - WISE
# Split students into two parts
part1 = students.iloc[:25]
part2 = students.iloc[25:]
combined_rows = pd.concat([part1,part2])
print(combined_rows.shape)

# 7. CONCAT COLUMN-WISE
student_basic = students[['Student_ID','Name']]
student_marks = marks[['Python']]
combined_cols = pd.concat([student_basic,student_marks],axis = 1)
print(combined_cols.head())

# 8. JOIN()
# Join() works using index
students_index = students.set_index("Student_ID")
contact_index = contact.set_index("Student_ID")
joined_df = students_index.join(contact_index)
print(joined_df.tail(10))
```

# PIVOT TABLE IN CSV

```python
# A pivot table is used to summarize and analyze data by grouping rows and applying aggregate functions such as sum, mean, count, max, min ,etc...
# STEP 1: Load Dataset
import pandas as pd
df = pd.read_csv("people-100.csv")
print(df.head())

#STEP 2: Create Age Column
df['Date of birth'] = pd.to_datetime(df['Date of birth'])
today = pd.Timestamp.today()
df['Age'] = ((today - df['Date of birth']).dt.days // 365)
print(df[['First Name','Sex','Age']].head())

# 1.Basic pivot table
#Average age by gender
pivot1 = pd.pivot_table(
    df,
    values='Age',
    index='Sex'
)
print(pivot1)

# 2. Pivot table with aggfunc = 'mean'
pivot2 = pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc='mean'
)
print(pivot2)
# Using sum
pivot2 = pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc='sum'
)
print(pivot2)
# Using Count, min, max
pivot2 = pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc='count'
)
print(pivot2)

# 3.Multiple Aggregate
pivot3 = pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc=['count','min','max']
)
print(pivot3)

# 4.Pivot table using margin
# Adds overall average
pivot4 = pd.pivot_table(
    df,
    values='Age',
    index='Sex',
    aggfunc='mean',
    margins=True
)
print(pivot4)

# 5. Group by Job Title and Gender
# Average age by Job Title and Gender
pivot5 = pd.pivot_table(
    df,
    values='Age',
    index='Job Title',
    columns='Sex',
    aggfunc='mean'
)
print(pivot5)

# Count people ny Job and Gender
pivot6 = pd.pivot_table(
    df,
    values='User Id',
    index='Job Title',
    columns='Sex',
    aggfunc='count',
    fill_value=0
)
print(pivot6)
# Nan fill value with 0
```
